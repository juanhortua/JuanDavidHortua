```python
import math

def fun(x):
    return (math.exp(-x))-x
    
def f(l,u):
    xl=l
    xu=u
    ite=1
    xrtemp=0
    
    while True:
        xr= ((xl+xu)/2.0)
        Ea= abs ((xr-xrtemp)/xr)*100
        print "{0:3d} | {1:5f} | {2:5f} | {3:5f}  | {4:5f}".format(ite,xl,xu,xr,Ea)
        
        if fun(xl)*fun(xr)<0.0:
            xu=xr
        elif fun(xl)*fun(xr)>0.0:
            xl=xr
        else:
            
            break   
            
        ite+=1
        
        xrtemp=xr
        
        
f(0.0,1.0)
```
